```python
#Combine Matplotlib with pandas to create advanced data visualizations with labeled data points and statistical overlays.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Sample Dataset: Sales performance
data = {
    "Salesperson": ["Alice", "Bob", "Charlie", "Diana", "Eve", "Frank", "Grace"],
    "Experience (Years)": [1, 3, 2, 5, 7, 4, 6],
    "Monthly Sales (K$)": [30, 50, 45, 70, 90, 65, 85],
}

# Create a pandas DataFrame
sales_df = pd.DataFrame(data)

# ----------------------------
# Scatter Plot with Labels
# ----------------------------
plt.figure(figsize=(10, 6))
plt.scatter(sales_df["Experience (Years)"], sales_df["Monthly Sales (K$)"], color="blue", s=100, label="Sales Data")

# Add labels for data points
for i, txt in enumerate(sales_df["Salesperson"]):
    plt.annotate(txt, (sales_df["Experience (Years)"][i], sales_df["Monthly Sales (K$)"][i]), 
                 textcoords="offset points", xytext=(5, 5), ha="center")

# ----------------------------
# Statistical Overlay: Linear Regression Line
# ----------------------------
# Fit a linear regression model
X = sales_df["Experience (Years)"].values.reshape(-1, 1)
y = sales_df["Monthly Sales (K$)"].values
model = LinearRegression().fit(X, y)
y_pred = model.predict(X)

# Plot regression line
plt.plot(sales_df["Experience (Years)"], y_pred, color="red", linestyle="--", label="Regression Line")

# ----------------------------
# Annotations for Outliers
# ----------------------------
# Mark high performers (e.g., Monthly Sales > 80K$)
high_performers = sales_df[sales_df["Monthly Sales (K$)"] > 80]
plt.scatter(high_performers["Experience (Years)"], high_performers["Monthly Sales (K$)"], color="green", s=150, edgecolor="black", label="High Performers")

for i, txt in enumerate(high_performers["Salesperson"]):
    plt.annotate(f"High: {txt}", (high_performers["Experience (Years)"].iloc[i], high_performers["Monthly Sales (K$)"].iloc[i]), 
                 textcoords="offset points", xytext=(0, -15), ha="center", color="green")

# ----------------------------
# Enhancements
# ----------------------------
plt.title("Sales Performance by Experience", fontsize=16)
plt.xlabel("Experience (Years)", fontsize=14)
plt.ylabel("Monthly Sales (K$)", fontsize=14)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()

# Show the plot
plt.show()
```
